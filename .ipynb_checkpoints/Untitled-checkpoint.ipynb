### KMeans
Kmeans (mÃ©thode des centre mobile) sert Ã  identifier des groupes d'observation ou clusters ayant des caractÃ©ristiques similaires

Ce schÃ©ma reprÃ©sente la minimisation de distortions(minimisation de la distance intra-groupe et maximisation de la distance inter-groupe)

```python

X2, y2 = make_classification(n_samples=1000, n_features=10, n_classes=2, random_state=42)
```

```python
distortions = []
K = range(1,15)
for k in K:
    kmeanModel = KMeans(n_clusters=k)
    kmeanModel.fit(X2)
    distortions.append(kmeanModel.inertia_)
    
plt.figure(figsize=(16,8))
plt.plot(K, distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('The Elbow Method showing the optimal k')
plt.show()
```

- On choisit K=2

```python
kmeans = KMeans(n_clusters=2)
kmeans.fit(X2)
pd.crosstab(y,kmeans.labels_)
```

La distribution de cluster 1 (default) est entre -2 et 2 et la distribution de cluster 0 (Pas de Default) est entre 1 et 3 pour la variable Historique => La variable Historique est significative et on peut l'utiliser pour segmenter les individus (les individus qui ont une mauvaise historique ont une tendance Ã  faire dÃ©fault)
Or la variable ValeurPropriÃ©tÃ© n'est pas significative pour la clustering kmeans

KNN
knn sert Ã  stocker tous les cas disponibles et classer les nouveaux cas en se basant sur une mesure de similaritÃ© (distance)

```python
error = []
# Calculer l'erreur pour k entre 1 et 40 
#Pour chaque itÃ©ration, l'erreur moyenne pour les valeurs prÃ©dites 
#de l'ensemble de test est calculÃ©e et sauvegardÃ©e ds la liste Erreur.
for i in range(1, 40):
    knn = KNeighborsClassifier(i)
    knn_model = knn.fit(X_train, y_train)
    pred_i = knn_model.predict(X_test)
    error.append(np.mean(pred_i != y_test))
plt.figure(figsize=(12, 6))
plt.plot(range(1, 40), error, color='red', linestyle='dashed', marker='o',
         markerfacecolor='blue', markersize=10)
plt.title('Taux Erreur pour les differentes valeurs de k')
plt.xlabel('K ')
plt.ylabel('Erreur')
```

On choisit k=4

```python
knn = KNeighborsClassifier(4)
knn_model = knn.fit(X_train, y_train)
y_pred_knn = knn_model.predict(X_test)
```

```python
print('Accuracy of K-NN classifier on training set: {:.2f}'
     .format(knn.score(X_train, y_train)))
print('Accuracy of K-NN classifier on test set: {:.2f}'
     .format(knn.score(X_test, y_test)))
```

```python
print(confusion_matrix(y_test, y_pred_knn))
```

 La premiÃ¨re classe reprÃ©sente les clients qui ne fait pas dÃ©fault et la deuxiÃ¨me classe reprÃ©sente les clients qui fait dÃ©fault

```python
print(classification_report(y_test, y_pred_knn))
```

Le modÃ¨le est globalement bon comme l'accuracy est Ã©gale Ã  0.76 (proche de 1)

La premiÃ¨re classe (Pas de dÃ©fault) est bien reprÃ©sentÃ© avec prÃ©cision, recall et f1-score proches de 1 or la deuxiÃ¨me classe (dÃ©fault) est mal reprÃ©sentÃ© avec une prÃ©cision de 0.37 (un grand nombre de positives fausses = 130), un recall de 0.37 (un grand nombre de nÃ©gatives fausses = 152) et un f1-score de 0.39

Cela peut Ãªtre expliquer par la grande diffÃ©rence de nombre entre les deux classe
