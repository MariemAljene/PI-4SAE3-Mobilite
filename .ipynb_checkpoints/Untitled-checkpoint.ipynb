### KMeans
Kmeans (méthode des centre mobile) sert à identifier des groupes d'observation ou clusters ayant des caractéristiques similaires

Ce schéma représente la minimisation de distortions(minimisation de la distance intra-groupe et maximisation de la distance inter-groupe)

```python

X2, y2 = make_classification(n_samples=1000, n_features=10, n_classes=2, random_state=42)
```

```python
distortions = []
K = range(1,15)
for k in K:
    kmeanModel = KMeans(n_clusters=k)
    kmeanModel.fit(X2)
    distortions.append(kmeanModel.inertia_)
    
plt.figure(figsize=(16,8))
plt.plot(K, distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('The Elbow Method showing the optimal k')
plt.show()
```

- On choisit K=2

```python
kmeans = KMeans(n_clusters=2)
kmeans.fit(X2)
pd.crosstab(y,kmeans.labels_)
```

La distribution de cluster 1 (default) est entre -2 et 2 et la distribution de cluster 0 (Pas de Default) est entre 1 et 3 pour la variable Historique => La variable Historique est significative et on peut l'utiliser pour segmenter les individus (les individus qui ont une mauvaise historique ont une tendance à faire défault)
Or la variable ValeurPropriété n'est pas significative pour la clustering kmeans

CAH
Classification Ascendante Hiérarchique sert à produire une structure (arborescence) permettant la mise en évidence de liens hiérarchiques entre les groupes d’individus et la détection d’un nombre de classes «naturel» au sein de la population

```python
Z = linkage(X2,'single')X
plt.figure(figsize=(12, 10))
plt.title('Classification Hiérarchique Ascendante(min)', fontsize=20)
d = dendrogram(Z)
plt.show()
```

```python
Z = linkage(X2, 'single')
plt.figure(figsize=(12, 10))
plt.title('Classification Hiérarchique Ascendante(min)', fontsize=20)
d = dendrogram(Z)
plt.show()
```

KNN
knn sert à stocker tous les cas disponibles et classer les nouveaux cas en se basant sur une mesure de similarité (distance)

```python
error = []
# Calculer l'erreur pour k entre 1 et 40 
#Pour chaque itération, l'erreur moyenne pour les valeurs prédites 
#de l'ensemble de test est calculée et sauvegardée ds la liste Erreur.
for i in range(1, 40):
    knn = KNeighborsClassifier(i)
    knn_model = knn.fit(X_train, y_train)
    pred_i = knn_model.predict(X_test)
    error.append(np.mean(pred_i != y_test))
plt.figure(figsize=(12, 6))
plt.plot(range(1, 40), error, color='red', linestyle='dashed', marker='o',
         markerfacecolor='blue', markersize=10)
plt.title('Taux Erreur pour les differentes valeurs de k')
plt.xlabel('K ')
plt.ylabel('Erreur')
```

On choisit k=4

```python
knn = KNeighborsClassifier(4)
knn_model = knn.fit(X_train, y_train)
y_pred_knn = knn_model.predict(X_test)
```

```python
print('Accuracy of K-NN classifier on training set: {:.2f}'
     .format(knn.score(X_train, y_train)))
print('Accuracy of K-NN classifier on test set: {:.2f}'
     .format(knn.score(X_test, y_test)))
```

```python
print(confusion_matrix(y_test, y_pred_knn))
```

 La première classe représente les clients qui ne fait pas défault et la deuxième classe représente les clients qui fait défault

```python
print(classification_report(y_test, y_pred_knn))
```

Le modèle est globalement bon comme l'accuracy est égale à 0.76 (proche de 1)

La première classe (Pas de défault) est bien représenté avec précision, recall et f1-score proches de 1 or la deuxième classe (défault) est mal représenté avec une précision de 0.37 (un grand nombre de positives fausses = 130), un recall de 0.37 (un grand nombre de négatives fausses = 152) et un f1-score de 0.39

Cela peut être expliquer par la grande différence de nombre entre les deux classe
